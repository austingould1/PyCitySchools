```python
import pandas
```

```python
#**District Summary**

#* Create a high level snapshot (in table form) of the district's key metrics, including:
 # * Total Schools
 # * Total Students
 # * Total Budget
 # * Average Math Score
 # * Average Reading Score
  #* % Passing Math
  #* % Passing Reading
  #* Overall Passing Rate (Average of the above two)
filepath = "raw_data/schools_complete.csv"
filepath_2 ="raw_data/students_complete.csv"
school_info = pandas.read_csv(filepath)
student_info = pandas.read_csv(filepath_2)
student_and_school = pandas.merge(student_info, school_info, how = "left", on=["school_name", "school_name"])
```

```python
student_and_school.head()
```

```python
# * Total Schools
school_count = len(student_and_school.school_name.unique())
school_count
```

```python
# * Total Students
total_student_count = student_and_school["Student ID"].count()
total_student_count
```

```python
# * Total Budget
total_budget = school_info["budget"].sum()
total_budget
```

```python
# * Average Math Score
average_math_score = student_and_school['math_score'].mean()
average_math_score
```

```python
# * Average Reading Score
average_reading_score = student_and_school['reading_score'].mean()
average_reading_score
```

```python
  #* % Passing Math
passing_math_count = student_and_school.loc[student_and_school['math_score'] >= 70].count()["Student ID"]
percent_passing_math = passing_math_count/total_student_count * 100
percent_passing_math
```

```python
#* % Passing Reading
passing_reading_count = student_and_school.loc[student_and_school['reading_score'] >= 70].count()["Student ID"]
percent_passing_reading = passing_reading_count/total_student_count * 100
percent_passing_reading
```

```python
#* Overall Passing Rate (Average of the above two)
total_passing_percent = (percent_passing_math + percent_passing_reading)/2
total_passing_percent
```

```python
#**School Summary**

#* Create an overview table that summarizes key metrics about each school, including:
  #* School Name
  #* School Type
  #* Total Students
  #* Total School Budget
  #* Per School Budget
  #* Average Math Score
  #* Average Reading Score
  #* % Passing Math
  #* % Passing Reading
  #* Overall Passing Rate (Average of the above two)
district_summary_df = pandas.DataFrame({
    "total schools": [school_count], 
    "total students": [total_student_count],
    "total budget": [total_budget],
    "average math score": [average_math_score],
    "average reading score": [average_reading_score],
    "% passing math": [percent_passing_math],
    "% passing reading": [percent_passing_reading],
    "overall passing rate": [total_passing_percent]
    
})
district_summary_df
```

```python
#* School Name
list_of_school_names = student_and_school.school_name.unique()
list_of_school_names
```

```python
#* School Type
# school_type = school_info[["school_name", "type"]]
# school_type
```

```python
school_types = school_info.set_index(["school_name"])["type"]
school_types
```

```python
#* Total Students
total_students_per_school = student_and_school.school_name.value_counts()
total_students_per_school
```

```python
#* Per School Budget
total_school_bud_per_school = student_and_school.groupby(["school_name"]).mean()["budget"]
total_school_bud_per_school
```

```python
per_school_capita = total_school_bud_per_school/total_students_per_school
per_school_capita
```

```python
#* Average Math Score
average_math_score = student_and_school.groupby(["school_name"]).mean()["math_score"]
average_math_score
```

```python
#* Average Reading Score
average_reading_score = student_and_school.groupby(["school_name"]).mean()["reading_score"]
average_reading_score
```

```python
#% passing math
passing_math_count = student_and_school[(student_and_school['math_score'] >= 70)]
percent_passing_math = passing_math_count.groupby(["school_name"]).count()["Student ID"]/total_students_per_school * 100
percent_passing_math
```

```python
#%passing reading
passing_reading_count = student_and_school[student_and_school['reading_score'] >= 70].groupby(["school_name"]).count()["student_name"]
percent_passing_reading = passing_reading_count/total_students_per_school * 100
percent_passing_reading
```

```python
#overall passing rate
overall_passing_rate =(percent_passing_reading + percent_passing_math)/2 
overall_passing_rate
```

```python
school_summary_df = pandas.DataFrame({
    "school type": school_types, 
    "total students": total_students_per_school,
    "total budget": total_school_bud_per_school,
    "average math score": average_math_score,
    "average reading score": average_reading_score,
    "% passing math": percent_passing_math,
    "% passing reading": percent_passing_reading,
    "overall passing rate": overall_passing_rate
    
})
school_summary_df
```

```python
#* Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

ninth_graders = student_and_school[student_and_school['grade'] == "9th"]
tenth_graders = student_and_school[student_and_school['grade'] == "10th"]
eleventh_graders = student_and_school[student_and_school["grade"]== "11th"]
twelfth_graders = student_and_school[student_and_school["grade"]== "12th"]
ninth_graders_scores = ninth_graders.groupby(["school_name"]).mean()["math_score"]
tenth_graders_scores = tenth_graders.groupby(["school_name"]).mean()["math_score"]
eleventh_graders_scores = eleventh_graders.groupby(["school_name"]).mean()["math_score"]
twelfth_graders_scores = twelfth_graders.groupby(["school_name"]).mean()["math_score"]
```

```python
school_math_per_school_df = pandas.DataFrame({
    "Ninth": ninth_graders_scores,
    "Tenth": tenth_graders_scores,
    "Eleventh": eleventh_graders_scores,
    "Twelfth": twelfth_graders_scores
})
school_math_per_school_df.index.name = None
school_math_per_school_df
```

```python
#* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
ninth_graders_scores = ninth_graders.groupby(["school_name"]).mean()["reading_score"]
tenth_graders_scores = tenth_graders.groupby(["school_name"]).mean()["reading_score"]
eleventh_graders_scores = eleventh_graders.groupby(["school_name"]).mean()["reading_score"]
twelfth_graders_scores = twelfth_graders.groupby(["school_name"]).mean()["reading_score"]
school_reading_per_school_df = pandas.DataFrame({
    "Ninth": ninth_graders_scores,
    "Tenth": tenth_graders_scores,
    "Eleventh": eleventh_graders_scores,
    "Twelfth": twelfth_graders_scores
})
school_reading_per_school_df.index.name = None
school_reading_per_school_df
```

```python
# * Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
#   * Average Math Score
#   * Average Reading Score
#   * % Passing Math
#   * % Passing Reading
#   * Overall Passing Rate (Average of the above two)

bins = [0,583,615,645,675]
labels = [">583", "583-615", "615-645", "645-685"]

pandas.cut(per_school_capita, bins=bins, labels=labels) 
school_summary_df['spending ranges']= pandas.cut(per_school_capita, bins=bins, labels=labels) 
school_summary_df 
spending_math_scores = school_summary_df.groupby(["spending ranges"]).mean()["average math score"]
spending_reading_scores = school_summary_df.groupby(["spending ranges"]).mean()["average reading score"]
spending_passing_math = school_summary_df.groupby(['spending ranges']).mean()['% passing math']
spending_passing_reading = school_summary_df.groupby(['spending ranges']).mean()['% passing reading']
overall_passing_spending = (spending_passing_math + spending_passing_reading)/2

spending_summary = pandas.DataFrame ({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "Overall Passing Rate": overall_passing_spending
})
spending_summary
```

```python
bins = [0,1000,2000,5000]
labels = [">small", "medium", "large"]

school_summary_df['school size']= pandas.cut(school_summary_df["total students"], bins=bins, labels=labels) 
school_summary_df 
spending_math_scores = school_summary_df.groupby(["school size"]).mean()["average math score"]
spending_reading_scores = school_summary_df.groupby(["school size"]).mean()["average reading score"]
spending_passing_math = school_summary_df.groupby(['school size']).mean()['% passing math']
spending_passing_reading = school_summary_df.groupby(['school size']).mean()['% passing reading']
overall_passing_spending = (spending_passing_math + spending_passing_reading)/2

size_summary = pandas.DataFrame ({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "Overall Passing Rate": overall_passing_spending
})
size_summary
```

```python
#**Scores by School Type**

#* Repeat the above breakdown, but this time group schools based on school type (Charter vs. District).
spending_math_scores = school_summary_df.groupby(["school type"]).mean()["average math score"]
spending_reading_scores = school_summary_df.groupby(["school type"]).mean()["average reading score"]
spending_passing_math = school_summary_df.groupby(['school type']).mean()['% passing math']
spending_passing_reading = school_summary_df.groupby(['school type']).mean()['% passing reading']
overall_passing_spending = (spending_passing_math + spending_passing_reading)/2


type_summary = pandas.DataFrame ({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "Overall Passing Rate": overall_passing_spending
})
type_summary
```
